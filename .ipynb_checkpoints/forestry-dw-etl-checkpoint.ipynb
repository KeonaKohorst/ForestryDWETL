```python
# work on image

# getting the columns figured out
images = pd.DataFrame(columns=[
    'sk_image_number', 'image_date', 'image_format', 'image_resolution',
    'image_scale', 'fly_height', 'focus_length', 'shutter_speed',
    'iso_sensitivity', 'lat_long_upper_right', 'lat_long_upper_left',
    'lat_long_lower_right', 'lat_long_lower_left'
])

images.head()
```

```python
#populate with what i can find for images from the dataset

import os
import pandas as pd
from PIL import Image
import numpy as np

# get image folder
image_folder = os.path.join(os.getcwd(), "Images") 

# empty list to store image data
image_data = []

# loop through the images in the folder
for idx, filename in enumerate(os.listdir(image_folder)):
    file_path = os.path.join(image_folder, filename)
    
    try:
        with Image.open(file_path) as img:
            image_resolution = f"{img.width}x{img.height}"  # resolution as "width x height"
            image_format = img.format  #  (JPEG, PNG, etc.)
            image_name = os.path.basename(img.filename)
            sk_image_number = idx + 1  # assign a surrogate key
            
            # Append data (fill non-available fields with NaN)
            image_data.append([
                sk_image_number,  # Unique ID
                image_name,
                np.nan,  # image_date (not available)
                image_format,  # Extracted from PIL
                image_resolution,  # Extracted from PIL
                np.nan,  # image_scale (not available)
                np.nan,  # fly_height (not available)
                np.nan,  # focus_length (not available)
                np.nan,  # shutter_speed (not available)
                np.nan,  # iso_sensitivity (not available)
                np.nan,  # lat_long_upper_right (not available)
                np.nan,  # lat_long_upper_left (not available)
                np.nan,  # lat_long_lower_right (not available)
                np.nan   # lat_long_lower_left (not available)
            ])
    except Exception as e:
        print(f"Error processing {filename}: {e}")

images = pd.DataFrame(image_data, columns=[
    'sk_image_number', 'image_name', 'image_date', 'image_format', 'image_resolution',
    'image_scale', 'fly_height', 'focus_length', 'shutter_speed',
    'iso_sensitivity', 'lat_long_upper_right', 'lat_long_upper_left',
    'lat_long_lower_right', 'lat_long_lower_left'
])

images.head()
```

```python
#fill for the fact table, connect images to species

columns = [
    "sk_tree_metrics_id", "fk_species_id", "fk_device_id", "fk_image_number",
    "capture_count", "tree_height", "tree_radius", "trunk_diameter",
    "confidence_score", "health_score", "year_planted",
    "tree_center_lattitude", "tree_center_longitude"
]

fact = pd.DataFrame(columns=columns)
fact.head()
```

```python
#populate the fk_species_id and fk_image_number

species_map = species.set_index("common_name")["sk_species_id"].to_dict()

print(species_map)
```

```python
dataset.head()
```

```python
fact_records = []
tree_metrics_id = 1 #surrogate key

# strip leading/trailing whitespace from dataset column names
dataset.columns = dataset.columns.str.strip()

for _, row in dataset.iterrows():
    
    #get the image number based on filename
    try:
        image_number = images.loc[images["image_name"] == row["filename"], "sk_image_number"].values[0]
    except IndexError:
        print(f"Error: Filename '{row['filename']}' not found in images DataFrame.")
        continue  #skip this row if no matching image number is found

    # iterate over species columns
    for species_name in species_map.keys():
        print("inside inner loop")
        print(f"Checking species column: {species_name}, value: {row[species_name]}") 
        print("here")
        # if the value in the species column is 1, we associate it with the species
        print(row)
        if row[species_name] == 1:
            print(f"Adding record for species: {species_name}, species_id: {species_map[species_name]}, image_number: {image_number}")  # Debugging print statement
            fact_records.append({
                "sk_tree_metrics_id": tree_metrics_id,
                "fk_species_id": species_map[species_name],  # species ID from the species map
                "fk_image_number": image_number,
                "fk_device_id": np.nan,
                "capture_count": np.nan,
                "tree_height": np.nan,
                "tree_radius": np.nan,
                "trunk_diameter": np.nan,
                "confidence_score": np.nan,
                "health_score": np.nan,
                "year_planted": np.nan,
                "tree_center_lattitude": np.nan,
                "tree_center_longitude": np.nan
            })
            tree_metrics_id += 1


fact = pd.DataFrame(fact_records)

fact.head()
```
